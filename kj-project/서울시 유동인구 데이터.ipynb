```python
dict2 = {"행자부행정코드":df['행자부행정동코드'][1], "시군구명":df['시군구명'][1],
        "행정동명":df['행정동명'][1]}
```

```python
dict1 = {}
dict1.update(dict2)
dict1
```

```python
from collections import defaultdict
dic = defaultdict(list)

for i in range(1, len(df)):
    dict1 = {"행정코드":df['행자부행정동코드'][i], "시군구명":df['시군구명'][i],
        "행정동명":df['행정동명'][i]}
    dic['행정분류'].append(dict1)
```

```python
for i in dic['행정분류']:
    print(i['행정코드'])
    print('---------------')
```

```python
if int(dic['행정분류'][0]['행자부행정코드']) == 11110515 :
    print(dic['행정분류'][0]['시군구명'], dic['행정분류'][0]['행정동명'])
```

```python
for i in Spop_local:
    code = i['ADSTRD_CODE_SE']
    print("-----"+code+"---------")
    for i in  dic['행정분류']:
        if i['행정코드'] == code :
            print(i['시군구명'], i['행정동명'])
            print("-----",i,"---------")
            
```
